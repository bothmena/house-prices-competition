```python
normalizers = {}
def normalize_column(X, column: str):
    df = X.loc[:, [column]]
    if column in normalizers.keys():
        normalizer = normalizers[column]
    else:
        normalizer = MinMaxScaler()
        normalizer.fit(df)
        normalizers[column] = normalizer
    np_scaled = normalizer.transform(df)
    X[column] = pd.DataFrame(np_scaled)[0]

for column in list(X_train.describe().columns):
    normalize_column(X_train, column)
X_train.head()
```
